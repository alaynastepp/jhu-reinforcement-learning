# Visualize metrics

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
metrics_arrays = [monte_metrics, sarsa_metrics, qlearning_metrics]
labels = ['Monte Carlo', 'SARSA', 'QLearning']
```

```python
plt.figure(figsize=(10, 6))
plt.bar(labels, [agent['avg_wins'] for agent in metrics_arrays], color=sns.color_palette("Set2", len(labels)))
plt.ylabel("Mean Winning Percentage (%)")
plt.title("Mean Winning Percentage Comparison")
plt.show()
plt.close()
```

```python
plt.plot(metrics.moving_average(monte_metrics['avg_rewards']), label='Monte Carlo', linewidth=2)
plt.plot(metrics.moving_average(sarsa_metrics['avg_rewards']), label='SARSA', linewidth=2)
plt.plot(metrics.moving_average(qlearning_metrics['avg_rewards']), label='QLearning', linewidth=2)
plt.xlabel("Episodes")
plt.ylabel("Mean Cumulative Return")
plt.title("Mean Cumulative Return over Episodes")
plt.legend()
plt.grid()
plt.show()
plt.close()
```

```python
visited_states = [np.sum(agent['state_action_visit_count'] > 0) / agent['state_action_visit_count'].size * 100 for agent in metrics_arrays]
plt.bar(labels, visited_states, color=sns.color_palette("Set2", len(labels)))
plt.ylabel("Mean Percentage of Visited States (%)")
plt.title("Mean Percentage of Visited States Comparison")
plt.show()
plt.close()
```

```python
# Plot each agent's winning percentage moving average
moving_avg = metrics.moving_average(monte_metrics['win_statuses'], window_size=30)
plt.plot(moving_avg, label="Monte Carlo")
moving_avg = metrics.moving_average(sarsa_metrics['win_statuses'], window_size=30)
plt.plot(moving_avg, label="SARSA")
moving_avg = metrics.moving_average(qlearning_metrics['win_statuses'], window_size=30)
plt.plot(moving_avg, label="QLearning")

plt.xlabel("Episodes")
plt.ylabel("Winning Percentage")
plt.title("Winning Percentage Over Episodes")
plt.legend()
plt.grid()
plt.ylim(0, 1.1)
plt.show()
plt.close()
```

```python
plt.plot(range(500), monte_metrics['avg_rewards'], label='MonteCarlo')
plt.plot(range(500), sarsa_metrics['avg_rewards'], label='SARSA')
plt.plot(range(500), qlearning_metrics['avg_rewards'], label='QLearning')
plt.ylabel("Average Rewards")
plt.title("Average Rewards over Training Comparison")
plt.legend()
plt.show()
plt.close()
```

```python
plt.plot(range(500), monte_metrics['avg_scores'], label='MonteCarlo')
plt.plot(range(500), sarsa_metrics['avg_scores'], label='SARSA')
plt.plot(range(500), qlearning_metrics['avg_scores'], label='QLearning')
plt.ylabel("Average Score")
plt.title("Average Score over Training Comparison")
plt.legend()
plt.show()
plt.close()
```
